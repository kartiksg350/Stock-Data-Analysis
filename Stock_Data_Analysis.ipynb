```python
# Importing necessary libraries
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data
# Load the datasets
part1_data = pd.read_csv('/Users/kartiksg/Downloads/PYTHON/Part1_data.csv')
part2_data = pd.read_csv('/Users/kartiksg/Downloads/PYTHON/Part2_data (1).csv')

# Display the first few rows of each dataset
print(part1_data.head())
print(part2_data.head())
```

```python
# Round all numerical columns in part2_data to two decimal places
part1_data = part1_data.round(2)

# Display the first few rows to verify
print(part1_data.head())

# Round all numerical columns in part2_data to two decimal places
part2_data = part2_data.round(2)

# Display the first few rows to verify
print(part2_data.head())
```

```python
#Since, the dataset is clean and has no missing or duplicate values. There is no need of data cleaning.
#Changing the format of date column to datetime format so that it can allow for proper time-series analysis.

# Convert the Date column to datetime in part1_data
part1_data['Date'] = pd.to_datetime(part1_data['Date'], format='%d-%m-%Y')

# Convert the Date column to datetime in part2_data
part2_data['Date'] = pd.to_datetime(part2_data['Date'], format='%d-%m-%Y')

# Verify the conversion by checking the data types
print(part1_data.info())
print(part2_data.info())
```

```python
#Now that date is changed to DateTime format. Let's PROCEED TO NEXT STEP OF DATA ANALYSIS.

#Question 1: Highest Closing Price? On which date were the highest closing prices observed in the provided stock data for Amazon?
# Find the maximum closing price
max_close_price = part1_data['Close'].max()

# Find all dates corresponding to the maximum closing price
max_close_dates = part1_data.loc[part1_data['Close'] == max_close_price, 'Date']

# Print the results
print(f"The highest closing price for Amazon was {max_close_price} and was observed on the following dates:")
print(max_close_dates)

                                                                                
```

```python
#Question 2: Positive Daily Returns

#Positive daily returns refer to the situation where the closing price of a financial asset, such as a stock, is higher at the end of a trading day compared to the opening price of that day. Positive daily returns indicate that the asset's price has increased during that particular trading day. They convey information about profitability, market sentiment, and the performance of individual assets or portfolios, helping investors and analysts make informed decisions in the financial markets.
# Step 1: Create a boolean column that indicates whether the daily return was positive
part1_data['Positive Return'] = part1_data['Close'] > part1_data['Open']

# Step 2: Count the number of days with positive returns
num_days_positive_returns = part1_data['Positive Return'].sum()

# Step 3: Get the total number of trading days
total_trading_days = len(part1_data)

# Step 4: Calculate the percentage of days with positive returns
percentage_positive_returns = (num_days_positive_returns / total_trading_days) * 100

# Step 5: Display the results
print(f"The number of days with positive daily returns is: {num_days_positive_returns} out of {total_trading_days} total trading days.")
print(f"This represents {percentage_positive_returns:.2f}% of the total trading days.")

```

```python
#Question 3: Daily Price Range

#The *daily price range* in stocks, also known as the trading range or intraday price range, is the difference between the highest price at which a stock is traded during a given trading day (the "high") and the lowest price at which it is traded during the same day (the "low").

#In mathematical terms: `Daily Price Range = High Price - Low Price`
#The daily price range provides valuable information to investors and traders about the volatility and price movement of a stock on a particular trading day.

# Step 1: Calculate the daily price range and create a new column
part1_data['Price Range'] = part1_data['High'] - part1_data['Low']

# Step 2: Find the lowest value in the Price Range column
lowest_price_range = part1_data['Price Range'].min()

# Print the lowest price range value
print(f"The lowest daily price range observed is: {lowest_price_range:.2f}")
```

```python
# Step 1: Calculate the percentage change in the closing price
part1_data['Price Change %'] = part1_data['Close'].pct_change() * 100

# Step 2: Display the first few rows to verify
print(part1_data[['Date', 'Close', 'Price Change %']].head())
```

```python
# Step 3: Count the number of days with a negative percentage change
negative_price_changes = part1_data[part1_data['Price Change %'] < 0].shape[0]

# Print the number of days with negative percentage changes
print(f"The number of days with a negative percentage change in closing price is: {negative_price_changes}")
```

```python
#Question 5: Total Trading Volume

#Note that the Volume column denotes the total number of shares of the company's stock that were traded on that day. Let's say you want to calculate the total trading volume for each year in the data set (2010-2020)

#Which year had the highest total trading volume?

# Step 1: Extract the year from the Date column
part1_data['Year'] = part1_data['Date'].dt.year

# Step 2: Group the data by year and calculate the total volume for each year
yearly_total_volume = part1_data.groupby('Year')['Volume'].sum()

# Step 3: Find the year with the maximum total volume
year_with_max_volume = yearly_total_volume.idxmax()
max_volume = yearly_total_volume.max()

# Print the results
print(f"The year with the highest total trading volume is: {year_with_max_volume} with a volume of {max_volume}")
```

```python
#Question 6: Average Closing Price for 2015

#Calculate the average closing price for the year 2015 (Round off to 2 decimal places).
# Step 1: Filter the data for the year 2015
data_2015 = part1_data[part1_data['Date'].dt.year == 2015]

# Step 2: Calculate the average closing price for 2015
average_closing_price_2015 = data_2015['Close'].mean()

# Step 3: Round the result to 2 decimal places
average_closing_price_2015 = round(average_closing_price_2015, 2)

# Print the result
print(f"The average closing price for the year 2015 is: ${average_closing_price_2015}")
```

```python
#Question 7 : Bar Chart

#Use a bar chart to find which month in 2015 had the least average closing price.

#The bar chart contains the following:
#Months along the X-axis
#Average closing price along the Y-axis
#Each bar should be labeled with the average closing price

# Step 1: Ensure the Date column is in datetime format (if not already done)
part1_data['Date'] = pd.to_datetime(part1_data['Date'], format='%d-%m-%Y')

# Step 2: Extract the year and month from the Date column
part1_data['Year'] = part1_data['Date'].dt.year
part1_data['Month'] = part1_data['Date'].dt.month

# Step 3: Filter the data for the year 2015
data_2015 = part1_data[part1_data['Year'] == 2015]

# Step 4: Group by month and calculate the average closing price
monthly_avg_closing_prices_2015 = data_2015.groupby('Month')['Close'].mean()

# Step 5: Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_avg_closing_prices_2015.index, y=monthly_avg_closing_prices_2015.values, palette="viridis")

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Average Closing Price')
plt.title('Average Monthly Closing Prices for Amazon in 2015')

# Label each bar with the average closing price
for index, value in enumerate(monthly_avg_closing_prices_2015.values):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom')

plt.show()

# Find the month with the least average closing price
min_avg_price_month = monthly_avg_closing_prices_2015.idxmin()
min_avg_price_value = monthly_avg_closing_prices_2015.min()

# Print the result
print(f"The month with the least average closing price in 2015 was: {min_avg_price_month} with an average closing price of ${min_avg_price_value:.2f}")
```

```python
#Question 8: Risk

#Risk in stocks refers to the uncertainty or variability in the returns that an investor may experience when holding a particular stock. It represents the possibility of both favorable and unfavorable outcomes, including the potential for losses. Understanding and quantifying risk is crucial for investors, as it helps them make informed investment decisions and manage their portfolios effectively.

#Standard deviation measures the dispersion or spread of a stock's returns around its average or expected return. A higher standard deviation indicates greater price volatility and is often used as a measure of risk.
#In which year did the Amazon stock have the highest level of risk in its closing price?

# Step 1: Ensure the Date column is in datetime format (if not already done)
part1_data['Date'] = pd.to_datetime(part1_data['Date'], format='%d-%m-%Y')

# Step 2: Extract the year from the Date column
part1_data['Year'] = part1_data['Date'].dt.year

# Step 3: Calculate daily returns (percentage change)
part1_data['Daily Return'] = part1_data['Close'].pct_change()

# Step 4: Group by year and calculate the standard deviation of daily returns
yearly_std_dev = part1_data.groupby('Year')['Daily Return'].std()

# Step 5: Identify the year with the highest risk (highest standard deviation)
year_with_highest_risk = yearly_std_dev.idxmax()
highest_std_dev_value = yearly_std_dev.max()

# Print the results
print(f"The year with the highest level of risk (highest standard deviation of daily returns) is: {year_with_highest_risk}")
print(f"The standard deviation of daily returns in that year was: {highest_std_dev_value:.6f}")
```

# Question 9: Candlestick Chart

## Candlestick charts are a valuable tool to visually interpret price data, understand market sentiment, and make trading decisions. They provide a concise and intuitive way to analyze historical price movements, identify potential opportunities, and manage risks in the financial markets.
Each candlestick consists of a rectangular "body" and two "wicks" or "shadows" extending from the top and bottom of the body. The body represents the price range between the opening and closing prices for a specific time period.
Stock-trading apps use candlestick charts extensively for analyzing trends in stock prices. These candlesticks are usually color-coded.

- A green or white candlestick typically indicates that the closing price was higher than the opening price, suggesting bullish (positive) sentiment.
- Conversely, a red or black candlestick represents a closing price lower than the opening price, signaling bearish (negative) sentiment.


## You can read more about candlestick charts from this [link](https://www.investopedia.com/articles/active-trading/062315/using-bullish-candlestick-patterns-buy-stocks.asp).

## Question 3: Correlation Heatmap

A correlation matrix is a square matrix that quantifies the pairwise relationships between variables in a dataset, with values ranging from -1 to 1, indicating the strength and direction of associations between the different variables

Heatmaps are a data visualization technique used to represent the distribution, density, or intensity of a phenomenon across a two-dimensional space. Heatmaps are typically presented as a grid of colored cells or rectangles, with each cell's color indicating the value of the underlying data at that specific location.

You can combine the above two and create a correlation heatmap to analyze the correlations between the stock prices of all the companies in a single view.

Create a correlation heatmap to determine which among the following pairs of stocks depict the highest correlation between them. Here are some additional instructions on how to build the heatmap

The heatmap should have a single color mapping
Different shades of the same color should indicate the value represented in each cell of the heatmap (the darker the shade, the higher the value)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Calculate the correlation matrix
correlation_matrix = data[['Close_AAPL', 'Close_AMZN', 'Close_GOOG', 'Close_MRK', 'Close_UNH']].corr()

# Step 2: Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=0.5)

# Add title
plt.title('Correlation Heatmap of Closing Prices')

# Display the plot
plt.show()
```
